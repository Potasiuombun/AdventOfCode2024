# Part1

```python
lists = open("lists.txt", "r").read()
```

```python
type(lists)
```

```python
splits = lists.splitlines()
```

```python
list1 = []
list2 = []
for x in splits:
    y = x.split("   ")
    
    l1 = int(y[0])
    l2 = int(y[1])
    list1.append(l1)
    list2.append(l2)
```

```python
list1, list2
```

```python
l1s = sorted(list1)
l2s = sorted(list2)
```

```python
l1s, l2s
```

```python
distances = []
for i,j in zip(l1s,l2s):
    n = abs(i-j)
    distances.append(n)
```

```python
sum(distances)
```

# Part 2

```python
similarity_score = []
for number in list1:
    score = number
    count = 0
    for second_number in list2:
        if number == second_number:
            count +=1
    score = score * count
    similarity_score.append(score)
```

```python
similarity_score[:10]
```

```python
sum(similarity_score)
```
